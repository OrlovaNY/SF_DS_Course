Определение: Функцией в математике называется правило, по которому можно сопоставить элементы двух множеств.

Формальное определение: Для любого элемента х из множества Х существует единственный элемент y из множества Y, такой что f(x) = y. Нельзя одному х поставить в соответствие два разных y. И в то же время двум х — один y: сколько угодно!

Числовая функция одной переменной —  это правило соответствия между числами x и y, согласно которому каждому x из области определения Df ставится в соответствие ровно один y из области значений Ef.

```python
#  импортируем необходимые библиотеки для работы с массивами чисел и для визуализации
import numpy as np
import matplotlib.pyplot as plt

# х - аргумент в нашей функции. 
# передадим начало и конец числового отрезка нашей функции и шаг аргумента:
x = np.arange(-10, 10.01, 0.01) # массив значений аргумента

# построение графика параболы
plt.plot(x, x**2) 

# включим сетку на графике 
plt.grid(True)

# показать график
plt.show() 
```

Элементарные функции — это функции, которые являются суммой, произведением или композицией элементов трех видов:

Степенные функции
Тригонометрические функции
Показательные и логарифмические функции

Элементарные функции можно записать понятной формулой из элементов–кирпичиков. А это значит, что их легко анализировать, зная свойства этих кирпичиков

В тригонометрии arctg (x) — это угол, тангенс которого равен х. То есть arctg (x) — функция обратная tg (x).
Возрастает внутри «коридора» (–π/2, π/2)
Вблизи нуля «похожа» на прямую y = x

Производная — есть предел приращений и тангенс угла наклона касательной. Задача о приближении, с помощью которой мы только что ввели понятие производной, называется линеаризацией.
