```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('tips.csv')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'], bins = 20, color = 'green')
axes.set_title('Общая сумма счета')
axes.set_ylabel('Количество заказов')
axes.set_xlabel('Сумма заказа, USD')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(x = df['total_bill'], y = df['tip'], marker = 'o')
axes.set_title('Общая сумма счёта Vs Cумма чаевых')
axes.set_xlabel('Сумма чека')
axes.set_ylabel('Размер чаевых')
```

```python
bills_per_day = df.groupby(by = 'day')['total_bill'].agg(['min','max']).reset_index()
```

```python
fig = plt.figure()

main_axes = fig.add_axes([0, 0, 1, 1])
main_axes.hist(df['total_bill'])

insert_axes = fig.add_axes([0.6, 0.6, 0.38, 0.38])
insert_axes.bar(x = bills_per_day['day'], height = bills_per_day['max'])
insert_axes.bar(x = bills_per_day['day'], height = bills_per_day['min'])
```

```python
fig = plt.figure()

main_axes = fig.add_axes([0, 0, 1, 1])
main_axes.hist(df['total_bill'])
main_axes.set_ylim(0,70)

insert_axes = fig.add_axes([0.5, 0.5, 0.48, 0.48])
insert_axes.set_ylim(0,70)# увеличение высоты графика
insert_axes.bar(x = bills_per_day['day'], height = bills_per_day['max'], 
                width = 0.4, align = 'edge', label='max')#label для легенды
insert_axes.bar(x = bills_per_day['day'], height = bills_per_day['min'], 
                width = -0.4, align = 'edge', label='min')
insert_axes.legend(loc=1)# 1 - 'upper right'
```

```python
# Создать несколько графиков
fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].hist(df['total_bill'], bins = 20, color = 'green')
axes[0].set_title('Общая сумма счета')

axes[1].hist(df['tip'], bins = 20, color = 'orange')
axes[1].set_title('Общая сумма чаевых')

axes[2].scatter(x = df['total_bill'], y = df['tip'], marker = '1')
axes[2].set_title('Общая сумма счёта Vs Cумма чаевых')
```

```python
import numpy as np 
```

```python
x = np.linspace(start=-3., stop=3., num=1000)
y1 = np.exp(x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y1) 
# Первым аргументом подается последовательность чисел аргумента, 
# вторым — последовательность значений функции для каждого аргументa
```

```python
y2 = 10 + 0.4*x - 0.3*x**2 + 0.1*x**3

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[300:], y1[300:]) #axes.plot(x, y1) 
axes.plot(x, y2)
```

```python
fig = plt.figure(figsize=(7, 4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y1, linestyle='--')
```

```python
# методом форматирования линии. Его структура — это строка, содержащая в полном виде символы, 
# обозначающие цвет маркера, тип маркера и тип линии. Возможен пропуск любого количества из этих трёх параметров. 
# Мы отрисуем линию в фиолетовом цвете ('m' — magenta) с маркером ромба ('D' — diamond marker):
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[::40], y2[::40], 'mD')
```
